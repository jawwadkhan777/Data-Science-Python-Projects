<h1><u><b>HANGMAN GAME</b></u></h1>
<p>Hangman is a word-guessing game usually played with two people (in this project it will play between computer and the user). One thinks of a word and draws a set of underscores, each representing a letter in the word. The other person tries to guess the word by suggesting letters one at a time. If a guessed letter is in the word, it's revealed in its correct position; if not, a part of a stick figure (representing a hangman) is drawn. The guessing continues until the word is guessed correctly or the hangman is fully drawn. The goal is to guess the word before the hangman is complete.</p>

```python
print("-------HANGMAN GAME-------")
```

```python
import random;
word_list = ["aardvark","baboon","camel"]
chosen_word = random.choice(word_list)

display = []
for letter in chosen_word:
  display.append("-")
print("Guessing sheet=>",display)

lives = 6
print("Total lives =>", lives)

stages = ['''
   +---+
   |   |
   O   |
  /|\  |
  / \  |
       |
  =========
  ''', '''
   +---+
   |   |
   O   |
  /|\  |
  /    |
       |
  =========
  ''', '''
   +---+
   |   |
   O   |
  /|\  |
       |
       |
  =========
  ''', '''
   +---+
   |   |
   O   |
  /|   |
       |
       |
  =========
  ''', '''
  +---+
  |   |
  O   |
  |   |
      |
      |
  =========
  ''', '''
  +---+
  |   |
  O   |
      |
      |
      |
  =========
  ''', '''
  +---+
  |   |
      |
      |
      |
      |
  =========''']
print("Your current stage: ", stages[lives])

while(True):
  guess = input("Guess a letter: ").lower()
  if guess in display:
    guess = input("'"+guess+"' "+"already guessed, guess another letter: ").lower()
  for place in range(len(chosen_word)):
    if guess==chosen_word[place]:
      display[place] = guess

  if "-" not in display:
    print(display)
    print("Hurrah, you won!!!")
    print("-------Game end, Thanks for playing Hangman!!!-------")
    break

  if guess not in chosen_word:
    lives -= 1
    print("'"+guess+"' "+"incorrect guess, lives remainig =>", lives)
    print("Now your current stage: ", stages[lives])
    if lives==0:
      print("Oh, you lose!!!")
      print("Correct guessing is:", chosen_word)
      print("-------Game end, Thanks for playing Hangman!!!-------")
      break
  print(display)
  print("----------------------------------------------------")
```
